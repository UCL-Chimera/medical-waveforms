```python
import os, sys #The first three cells are from Finn's example notebook. 
#The next cells generates the power spectrum of the exmaple aerterial waveform
sys.path.append(os.pardir)  # Needed for sidewinder import before we install it

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sidewinder.fouriertools 
from sidewinder.utils import get_root_directory
```

```python
waveforms = pd.read_csv(
    os.path.join(
        get_root_directory(),
        'local_data',
        'abp_waveform_extract.csv'
    ),
    index_col=0
)

art = waveforms['SNUADC/ART'].values[100205:101980]
```

```python
fig, ax = plt.subplots(figsize=(14, 3), constrained_layout=True)
ax.plot(art)
ax.set(xlabel='Sample index', ylabel='Arterial pressure (mmHg)')
plt.show()
```

```python
xp,yp=sidewinder.fouriertools.powerspec(art,fs=125,ns=4)
```

```python
plt.plot(xp,yp)
plt.yscale('log')
plt.xlabel('Frequency(Hz)')
plt.ylabel('Power')
plt.show()
```
